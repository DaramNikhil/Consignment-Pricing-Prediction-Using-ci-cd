```python
def Remove_Unwanted_data_Features(data_feature):
    try:
        data_feature = data_feature.drop("PQ #", axis = 1)
        data_feature = data_feature.drop("ASN/DN #", axis = 1)
        data_feature = data_feature.drop("PO / SO #", axis = 1)
        data_feature = data_feature.drop("Dosage Form", axis = 1)
        data_feature = data_feature.drop("Molecule/Test Type", axis = 1)
        data_feature = data_feature.drop("Vendor", axis = 1)
        data_feature = data_feature.drop("Manufacturing Site", axis = 1)
        data_feature = data_feature.drop("Project Code", axis =1 )
        data_feature = data_feature.drop("ID", axis = 1)
        return data_feature
    except Exception as e:
        print("Error in {}".format(e))
        raise e
```

```python
data_1 = Remove_Unwanted_data_Features(data_feature = data)
data_1.head()
```

```python
def Handling_Missing_Data(data_feature):
    try:
        data_feature["Shipment Mode"] = data_feature["Shipment Mode"].fillna('Air')
        data_feature["Line Item Insurance (USD)"] = data_feature["Line Item Insurance (USD)"].fillna(data_feature["Line Item Insurance (USD)"].median())
        data_feature = data_feature.drop("Dosage", axis = 1)
        return data_feature
    except Exception as e:
        print("error in {}".format(e))
        raise e
    
```

```python
class Feature_Engineering_Configure_Class:
    try:
        def __init__(self, data_feature) -> None:
            self.data_feature = data_feature

        def PQ_First_Sent_to_Client_Date_func(self, shipment_data_f):
            if shipment_data_f == "Pre-PQ Process":
                return pd.to_datetime('01/06/2009', format="%d/%m/%Y")
            elif shipment_data_f == "Date Not Captured":
                return "Date Not Captured"
            
            else:
                if len(shipment_data_f) < 9:
                    shipment_data_f = pd.to_datetime(shipment_data_f, format="%m/%d/%y")
                    return shipment_data_f
                else:
                    shipment_data_f = shipment_data_f.replace("-", "/")
                    shipment_data_f = pd.to_datetime(shipment_data_f, format="%d/%m/%Y")
                    return shipment_data_f
                
        def PQ_First_Sent_to_Client_Date(self):
            try:
                self.data_feature["PQ First Sent to Client Date"] = self.data_feature["PQ First Sent to Client Date"].apply(self.PQ_First_Sent_to_Client_Date_func)
                return self.data_feature
            except Exception as e:
                print("Error in {}".format(e))
                raise e

    
    except Exception as e:
        print("Error in {}".format(e))
        raise e
```
